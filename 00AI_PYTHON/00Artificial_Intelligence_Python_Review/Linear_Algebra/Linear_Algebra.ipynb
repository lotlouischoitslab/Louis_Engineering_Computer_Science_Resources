```python
import numpy as np
import sympy as syms
from sympy import *
```

$$ 3x - y = 7$$
$$ 2x + y = 8$$

```python
x = symbols('x')
y = symbols('y')
```

```python
eq1 = 3*x-y-7
eq2 = 2*x+y-8
linsolve([eq1,eq2],(x,y))
```

```python
def convert_to_radians(theta):
    return ((45/180)*np.pi)
```

```python
t1 = symbols('t1')
t2 = symbols('t2')
g = 9.81
mass = 25
theta_deg = 45
theta = convert_to_radians(theta_deg)
```

```python
eq1 = t1-t2*np.cos(theta)
eq2 = (t1*np.sin(theta))-(mass*g)
linsolve([eq1,eq2],(t1,t2))
```
