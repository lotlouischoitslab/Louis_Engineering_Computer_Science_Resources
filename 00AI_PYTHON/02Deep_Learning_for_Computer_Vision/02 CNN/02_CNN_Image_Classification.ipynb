```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import os
import random
os.chdir('C:/Users/LOTLOUISCHO/Desktop/02MyProjects/ktx_project')
from tensorflow.keras.preprocessing.image import load_img
```

```python
#0 means KTX-1
#1 means KTX-Sancheon
#2 means KTX-EUM
#3 means SRT
input_path = []
label = []

for classes in os.listdir('trains'):
    for path in os.listdir('trains/'+classes):
        if classes == 'ktx_1':
            label.append(0)
        elif classes == 'ktx_sancheon':
            label.append(1)
        elif classes == 'ktx_eum':
            label.append(2)
        elif classes == 'srt':
            label.append(3)
        input_path.append(os.path.join('trains',classes,path))
```

```python
df = pd.DataFrame()
df['images'] = input_path
df['label'] = label
df = df.sample(frac=1).reset_index(drop=True)
df.head()
```

## Exploratory Data Analysis

```python
#display grid of images
plt.figure(figsize=(25,35))
temp = df[df['label']==0]['images']
start = random.randint(0,len(temp))
files = temp[start:start+25]

for index,file in enumerate(files):
    plt.subplot(5,5,index+1)
    img = load_img(file)
    img = np.array(img)
    plt.imshow(img)
    plt.title('KTX-1')
    plt.axis('off')
```
