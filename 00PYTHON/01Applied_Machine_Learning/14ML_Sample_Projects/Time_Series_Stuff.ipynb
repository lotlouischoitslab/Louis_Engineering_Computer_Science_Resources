```python
import numpy as np  
import pandas as pd
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
df = pd.read_csv('https://bit.ly/udemy_df',index_col='Date',parse_dates=True).asfreq('D')
df.head()
```

# Holt-Winters

```python
model = ExponentialSmoothing(endog=df.Udemy,trend='add',seasonal='add',seasonal_periods=7).fit()
```

```python
predictions = model.forecast(steps=150)
```

```python
df['Udemy']['2020-01-01':].plot(figsize=(9,6))
predictions.plot()
```
