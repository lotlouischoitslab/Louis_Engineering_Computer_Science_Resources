```python
def locate_card(cards,query):
    low,high = 0, len(cards)-1
    
    while low <= high:
        mid = (low+high)//2
        mid_elem = cards[mid]
        
        if mid_elem == query:
            return mid
        elif mid_elem < query:
            high = mid - 1
        elif mid_elem > query:
            low = mid + 1
        
    return -1 
```

```python
my_list = [1,2,3,4,5,6,7,8,9]
find = 5
index = locate_card(my_list,find)
print(index)
```

```python
def locate_card_rec(arr,low,high,query):
    if high >= low:
        mid = (low + high) //2
        
        if arr[mid] == query:
            return mid
        
        elif arr[mid] > query:
            return locate_card_rec(arr,mid - 1,high,query)
        else:
            return locate_card_rec(arr,low, mid + 1,query)
    else:
        return -1
```

```python
my_list = [1,2,3,4,5,6,7,8,9]
find = 5
low = 0
high = len(my_list)
index = locate_card_rec(my_list,low,high,find)
print(index)
```

### Just for review Linear Search

```python
def linear_search(arr,query):
    for i in range(len(arr)):
        if arr[i] == query:
            return (i)
    return -1
```

```python
my_list = [1,2,3,4,5,6,7,8,9]
find = 5
index = linear_search(my_list,find)
print(index)
```

### Binary Search Practice

```python
def binarysearch(nums,target):
    left,right = 0,len(nums)-1
    while left <= right:
        mid = (left+right)//2
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid+1
        elif nums[mid] > target:
            right = mid-1
    return -1
```

```python
def binarysearchrec(left,right,nums,target):
    if left > right:
        return -1
    mid = (left+right)//2
    if nums[mid] == target:
        return mid
    elif nums[mid] > target:
        return binarysearchrec(left,mid-1,nums,target)
    else:
        return binarysearchrec(mid+1,right,nums,target)
```

```python
nums = [1,2,3,4,5,6,7,8,9]
target = 6
iterate = binarysearch(nums,target)
recurse = binarysearchrec(0,len(nums)-1,nums,target)
print(iterate)
print(recurse)
```

##### Rotate a List

```python
def count_rotations(nums):
    positions = 0
    
    while positions < len(nums):
        if positions > 0 and nums[positions] < nums[positions - 1]:
            return positions
        positions +=1
    return 0
```

```python
list = [5,6,9,0,2,3,4]
nums = count_rotations(list)
print(nums)
```

```python
list = [0,2,3,4,5,6,9]
nums = count_rotations(list)
print(nums)
```
