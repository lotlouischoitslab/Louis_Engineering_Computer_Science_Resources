```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
# import heapq

# class Solution:
#     def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
#         merged = ListNode(-1)
#         curr = merged
#         vals_to_lists = [] #O(klogk) insert k elements into min heap without heapification.
        
#         for list in lists:
#             if list:
#                 val_to_lists = (list.val,list)
#                 vals_to_lists.append(val_to_lists)
        
#         myheap = heapq.heapify(vals_to_lists)
#         for i in range(len(vals_to_lists)):
#             print(vals_to_lists[i])
            
#         while len(vals_to_lists) != 0:
#             heapq.heappop(vals_to_lists)
#             min_ = set()
#             merged.next = min_.pop()
#             merged = merged.next
#             min_.second = min_[1].next
#             if (min_.second):
#                 val_to_list = (min_[1].val,min_[1])
#                 vals_to_lists.append(val_to_list)
#                 heapq.push_heap(vals_to_lists)
        
#         return merged.next

class Solution:
    #Divide-and-conquer approaches
    def mergeTwo(self,l1,l2):
        merged = ListNode(-1)
        merged_ptr = merged
        l1_ptr = l1
        l2_ptr = l2
        while(l1_ptr or l2_ptr):
            if l1_ptr is None:
                merged.next = l2_ptr
                break
            elif l2_ptr is None:
                merged.next = l1_ptr
                break
            else:
                if l1_ptr.val < l2_ptr.val:
                    merged_ptr.next = l1_ptr
                    
                    l1_ptr = l1_ptr.next
                else:
                    merged_ptr.next = l2_ptr
                    l2_ptr = l2_ptr.next
                    
                merged_ptr = merged_ptr.next
        return merged.next
            
    
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        if len(lists) == 0:
            return None
        if len(lists) == 1:
            return lists[0]
        
        num_lists = len(lists)
        
        while(num_lists > 1):
            for i in range(0,len(lists),2):
                if i + 1 < num_lists:
                    lists[i] = self.mergeTwo(lists[i], lists[i+1])
                else:
                    lists[i] = self.mergeTwo(lists[i], None)
                    
            for i in range(0,len(lists),2):
                if i + 1 < num_lists:
                    lists.pop(i+1)
                
            num_lists = len(lists)
        return lists[0]
```
