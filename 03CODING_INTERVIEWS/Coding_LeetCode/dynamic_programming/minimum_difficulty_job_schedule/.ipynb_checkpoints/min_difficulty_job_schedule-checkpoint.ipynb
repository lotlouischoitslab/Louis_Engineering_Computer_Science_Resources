# Algorithm VERY HARD PROBLEM!
State variables: We have a multidimensional DP problem, so we will have more than one state variable.
In any scenario, the minimum pieces of information that we will need to know is the index i of the job that we will start on for any particular day day. So our two state variables are i, which is an index into the vector jobDifficulty where jobDifficulty size represents the number of jobs that we can schedule for all of the d days. Our next state variable day day, which is not 0-based like i and represents the exact day starting from day 1. 

Not specific to any approach and execute this segment of code before doing anything else. As/Sanity check/Base case, we can return -1 whenever n < d since we know that we need to finish at least one task for each of the d days but we don't have enough tasks when the total number n is less than the number of days d for that job schedule. 

Algorithm Explanation: 
We know that we must finish at least 1 task each day, so starting at index i for any particular day, the very last inclusive index of the job that we can reach for that day is n - (d - day) - 1 where d - day gives us the number of days remaining, and subtracting (d - day) from n will give us the index into the array of jobDifficulty where we may begin for the next day if we decide that we should only complete 1 job for each of the remaining days. By subtracting 1, we have the very last index for which we should call the recurrence relation on. 

Notice that we are guaranteed at least 1 job for the job schedule and at least 1 day over which we can schedule the jobs. We will maintaing a variable for the hardest job we have encountered thus far starting from index i all the way up to index jobDifficulty size - n - (d - day) - 1 for every index in this inclusive range, we will update the hardest by taking the max between its current value or the value for jobDifficulty at the current index at the range. We will add the hardest job for the day with the result of the recursive call for our recurrence relation. So, there will be some work to compute the minimum difficulty of the job schedule for each stat i and day, and we will need to multiply this work by the number of states in order to analyze the runtime. 

Top-Down Approach (Memoization)
1. A function or data structure such as an array to compute/hold the answer to the problem for every state.
We will have a function dp that returns the minimum difficulty of a job schedule starting from index i (in the array jobDifficulty) for the day day. We know that the difficulty of a day is the maximum difficult of a job done on that day. However we don't know how many jobs we will complete on that day. All we know is that we want to minimize the difficulty of the job schedule starting from index i. So, as we indicated above, we will need to compute the maximum difficulty job that we have encountered thus far until we reach the final exclusive index that we can for that day of n - (d - day). We will also cacher our intermediate results in a hashmap, so that we can perform a lookup in O(1) time without having to re-calculate our results using the recurrence relation whenever encountering a duplicate input for the starting index of the job i and the day day.

2. A recurrence relation to transition between states.
Notice that we want the minimum difficulty of a job schedule starting from index i going until index n - (d - day) - 1 (inclusive). Therefore, we will take the minimum of the hardest job we have encountered thus far and add it to the result of the recursive call. We expect each of the previous days to have returned the minimum difficulty of their job schedule starting from the index i they were passed in as input, so we only ever have to worry about the current and the future days. So, we will iterate starting from index n - (d - day) (exclusive) and we will donate the loop variable as j. For index j, we can maintain the hardest job we have encountered thus for by storing the maximum job difficult starting from index i until index j and only updating it when we find a jobDifficulty[j] that exceeds the current maximum. We will add this to the return value of the recursive call dp(j + 1, day + 1) since we assume that j is the last inclusive index for the job that we will perform this day so j + 1 becomes the starting index for the job we will need to complete the next day. We increment day to indicate that we are moving on to the next day. Note that we want the minimum of the hardest job + dp(j + 1, day + 1) for all j since this would give us the minimum difficulty of job schedule for the current and the future days. 

Mathematically, our recurrence relation is:
Minimum(hardest + dp(j + 1, day + 1)) for all i <= j <= n - (d - day) - 1 where hardest is max(jobDifficulty[k]) for all i <= k <= j. 

3. Base Case to stop recursion
If d == day, then we are finished. We have no more days over which we can distirbute the remaining jobs, so we can return the difficulty of the present day, which is simply the maximum difficulty of the job done on day d. So starting from index i, until index n - 1, we will check for the maximum value in the array jobDifficulty and then return it. We can also pre-compute the hardest job remaining for each index i so that when the base case is reached, we don't have any extra work involved with finding the max starting from state variable i until the end of the array jobDifficulty representing the difficulty of the very last day, which is the max difficulty among all jobs done on that day. We will do this outside of our recursive function by mainting the hardest job done until index j starting from the end. and setting the corresponding array entry at index j with this value. 

Runtime Analysis:
Space Complexity: Notice that we won't actually visit every state. With the top-down approach, we will explicitly follow the constraint that we have to finish at least one task everyday. So let's consider an example: we have 10 jobs that we have to distribute over 5 days. LEt's consdier the second day, the smallest index we can start at is 1 if we were to start at the 0th index on day 1 and only complete 1 taks for that day. Then, on the second day, the very last index of the job we can complete for this day is 6. Subtracting 1 from 6, we get 5, which is n - d (also can be seen from the mathematical expression for the very last index we can visit: n - (d - day) - 1). 
#We notice that for each of the d days, we will make n - d recursive calls. Therefore the maximum number of recursive calls that our recursion stack will need to store is (n - d) * d or the total number of states for the top-down approach. Once we hit the base case, we will execute these recursive calls in LIFO order. Therefore our overall runtime would be O((n-d)*d).


Bottom-Up Approach (Tabulation)
1. A function or data structure to compute/hold the answer to the problem for every state
#We will maintain a 2d array dp where the size of the second dimension of d + 1 (to account for the base case where d == day where day is the index into the second dimension of dp) and the size of the first dimension will be the total number of jobs we can schedule or the size of the array jobDifficulty. So, now we will need to translate our recursion approach to an iteration approach. We will convert all our recursive function calls into array lookups where we index into the first dimension with i and index into the second dimension with day. The last column of our 2d array dp will become the hardestJobRemaining array from our top-down approach (i.e store its values). Our outer loop will iterate backwards over the size of the array jobDifficulty and our inner loop will iterate backwards from the total number of days d over which we can schedule jobs to 1. Notice that we need to iterate backwards since in our top-down approach, we needed to wait for the subproblem results from the recursive call before figuring out the result to our larger, overarching problem (i.e for the current day and starting at index i for the job). However, now we don't have the liberty of waiting for our base case to be hit before executing our recursive calls in LIFO order, so instead we need to compute the results of our subproblems before even reaching our current problem (as evident by the incrementing of day by 1 and incrementing of j by one for the second argument and first argument, respectively, in our recursive call). #Note inside of our nested loop, we will need another loop that is the same as the loop with variable and index j from our top-down approach 


Space Complexity: We have a 2d array with a size of n in the first dimension where n represents the total number of jobs that we can schedule over d days and the size of the second dimension is d + 1 Also, the space complexity is proportional to the number of states and each entry of our 2d array dp represents a particular state for the starting index i of a job corresponding to the day day. Therefore, our overall memory usage is O(n*d).

Runtime Analysis overall: We know from the bottom-up approach that we have a total of (n - d)*d states. For each state, we now need to consider the extra work required to compute the max difficulty of the job schedule starting from index i for the array of jobDifficulty and on day day. We have an additional for loop inside the body of our recursive function in the case of the top-down approach and inside the nested for loop in the case of our bottom-up approach. In the worst case scenario (i.e when the starting index i i s 0 for the job and it's the first day for the job schedule), we iterate n - d times. So, in the worst case, the extra work required for the states is O(n-d). Therefore, our overall runtime would is O(d*(n-d)^2).  

```python
#Bottom-Up Approach
class Solution:
    def minDifficulty(self, jobDifficulty: List[int], d: int) -> int:
        n = len(jobDifficulty)
        # If we cannot schedule at least one job per day, 
        # it is impossible to create a schedule
        if n < d:
            return -1
        
        dp = [[float("inf")] * (d + 1) for _ in range(n)]
        
        # Set base cases
        dp[-1][d] = jobDifficulty[-1]

        # On the last day, we must schedule all remaining jobs, so dp[i][d]
        # is the maximum difficulty job remaining
        for i in range(n - 2, -1, -1):
            dp[i][d] = max(dp[i + 1][d], jobDifficulty[i])

        for day in range(d - 1, 0, -1):
            for i in range(day - 1, n - (d - day)):
                hardest = 0
                # Iterate through the options and choose the best
                for j in range(i, n - (d - day)):
                    hardest = max(hardest, jobDifficulty[j])
                    # Recurrence relation
                    dp[i][day] = min(dp[i][day], hardest + dp[j + 1][day + 1])

        return dp[0][1]
```

```python
Top-Down Approach (Memoization)
class Solution:
    def minDifficulty(self, jobDifficulty: List[int], d: int) -> int:
        if len(jobDifficulty) < d:
            return -1
        
        min_difficulties = dict()
        hardest_job_remaining = [0]*(len(jobDifficulty))
        hardest = float('-inf')
        for j in range(len(jobDifficulty)-1,-1,-1):
            hardest = max(hardest,jobDifficulty[j])
            hardest_job_remaining[j] = hardest
        
        def dp(i,day, jobDifficulty, min_difficulties,d,hardest_job_remaining):
            if day == d:
                return hardest_job_remaining[i]
            
            if ((i,day) in min_difficulties):
                return min_difficulties[(i,day)]
            
            hardest = float('-inf')
            min_difficulty = float('inf')
            
            for j in range(i,len(jobDifficulty) - (d - day)):
                hardest = max(hardest,jobDifficulty[j])
                min_difficulty = min(min_difficulty, hardest + dp(j+1,day+1,jobDifficulty, min_difficulties,d,hardest_job_remaining))
            
            min_difficulties[(i,day)] = min_difficulty
            return min_difficulties[(i,day)]
                
        return dp(0,1,jobDifficulty,min_difficulties,d,hardest_job_remaining)
        
```

```python
#Optimized Tabulation Approach
#Notice that we only really ever need the column to the right of us in our 2d array dp (i.e we only ever need the minimum difficulty of the job schedule for the next day but for every starting index i of the job). Therefore, we will need a 1d array with the same size as the first dimension of array dp (n, or jobDifficulty size) and we can do away with the 2d array since we don't need the array values (or columns) for any of the other days except day + 1. The auxillary data structure we need is this array for the next day and the current day. The reason we need an array for the next day is so that at the end of our iteration of the outer loop, before we move on to the previous day since we're going backwards through the days, the minimum difficulty of the job schedule for the current day for each starting index i of the job will become the minimum difficulty of the job schedule of the previous day's next day. We will convert our 2d array value retrievals int o a 1d array value retrieval by indexing into the next day whenever we index into the 2d array dp at position day + 1 i the second dimension and indexing into the current day whenever we index into the 2d array dp at position day in the second dimension. So our overall memory usage is reduced from O(n*d) to O(n). Since we can ignore coefficients in Big-O expressions, this is simplifies into Big O(n) 
class Solution:
    def minDifficulty(self, jobDifficulty: List[int], d: int) -> int:
        n = len(jobDifficulty)
        # If we cannot schedule at least one job per day, 
        # it is impossible to create a schedule
        if n < d:
            return -1
        
        next_day = [0]*n
        hardest = float('-inf')
        for i in range(n-1,-1,-1):
            hardest = max(hardest,jobDifficulty[i])
            next_day[i] = hardest
        
        
        #current day is the next day of my previous day
        curr_day = [0]*n
        for day in range(d - 1,0,-1):
            for i in range(n - 1, -1, -1):
                hardest = float('-inf')
                min_difficulty = float('inf')
                # Iterate through the options and choose the best
                for j in range(i, n - (d - day)):
                    hardest = max(hardest, jobDifficulty[j])
                    min_difficulty = min(min_difficulty,hardest+next_day[j+1])
        
                curr_day[i] = min_difficulty
            next_day,curr_day = curr_day,next_day #Don't forget to swap
        return next_day[0]
```
