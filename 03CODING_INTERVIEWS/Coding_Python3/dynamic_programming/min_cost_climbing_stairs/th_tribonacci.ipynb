```python
# class Solution:
#     #Top Down Approach: Memoization
#     def tribonacci(self, n: int) -> int:
#         cache = dict()
#         def dp(n,cache):
#             if n == 0:
#                 return 0
#             if n == 1 or n == 2:
#                 return 1
#             else:
#                 if n not in cache:
#                     cache[n] = dp(n-1,cache) + dp(n-2, cache) + dp(n-3,cache)
#                 return cache[n]
#         return dp(n,cache)


class Solution:
    #Bottom Up Approach: Tabulation
#     def tribonacci(self, n: int) -> int:
#         if n == 0:
#             return 0
#         if n == 1:
#             return 1
        
#         dp = [0] * (n + 1)
#         dp[0] = 0
#         dp[1] = 1
#         dp[2] = 1
        
#         for i in range(3, n+1):
#             dp[i] = dp[i - 1] + dp[i - 2] + dp[i - 3] 
        
#         return dp[n]
    
    #Bottom-Up Approach without array
    def tribonacci(self, n: int) -> int:
        if n == 0:
            return 0
        
        
        prev3 = 0 
        prev2 = 1 
        prev1 = 1 
        
        for i in range(3,n + 1):
            temp1 = prev1 #Update to the right
            temp2 = prev2 #Update to the right
            prev1 = prev1 + prev2 + prev3  #prev1 gets updated
            prev2 = temp1 #Update to the right
            prev3 = temp2 #Update to the right
        
        return prev1
    
    
    
#State Variable: index n representing the original input n that we receive for function tribonacci

#1. A function/ data structure to compute/hold the answer to the problem for every state
#In the memoization approach, function dp will compute the Fibonnaci number for the input n (in terms of the original problem), or n - 1, n - 2. until the base cases for each sub problems

#In the tabulation approach, the array dp will store the results of the tribonacci numbers for each index i until we reach index n meaning our array will be the size (n + 1). 

#2. A recurrence relation to transition between the states. 
#Rearranging the recurrence relation we have above, we get T(n) = T(n - 1) + T(n - 2) + T(n - 3) for n >= 3. So, we don't use the recurrence relation when n < 3.

#3. Base Case(s) to stop recursion and not continue infinitely untill we reach an input of negative infinity. 

#Given above once again,
#T(0) = 0
#T(1) = 1
#T(2) = 1

#Runtime Analysis: O(n) where n represents the original input to the tribonacci function. In the memoization approach, for each unique inputs, we will make use of recurrence relation once. If we see the same input again, we will look to our cache for its tribonacci number in constant time and immediately return it. This does not require re-calculating the tribonacci number and thus re-using the recurrence relation. Therefore we will only use the recurrence relation for each unique input from n to base cases. In the tabulation approach, we have a for loop that iterates from i = 3 to n, so we do a traversal of our array dp. Therefore our overall runtime would be O(n) for both approaches.

#Space Complexity: O(n). In the case of memoization, we have a recursion stack that stores the recursive calls for each unique input from n down to the base case cases so that when the base cases are reached, we can stop recursing and then execute the recursive calls in the stack in LIFO order until we figure out he tribonacci number for input n. IN the tabulation approach, we have an auxiliary data structure (i.e array) with size n + 1 to store each of the tribonacci number corresponding to the array index. In both cases, our overall memory usage is O(n).

#Tabulation Approach without array
#An array requires linear space complexity, so we will entirely replace the array with couple of pointers. Notice that in every iteration of our loop, we only index our array dp at 0-based positions i - 1, i - 2 and i - 3. 
#Let's maintain 3 pointers to store these values. Prev one is going to store a tribonacci number at index of i - 1. Prev2 is going to store the tribonacci number of i - 2. prev3 is going to to store the tribonacci number of i - 3. 
#Every iteration of our loop, we will need to change prev 3 to hold prev 2's value and prev2 to hold prev1's value. Lastly, prev one need to store the tribonacci number of i which is simply the result of the recurrence relation. We will need temporary pointers in between to hold the tribonacci numbers at index i - 1 and index i -2 in order to successfully perform the update. 

#Space Complexity: O(1) We don't use any addiitonal auxiliary data structure. We simply maintain two pointers. We are going to store prev1 prev2 and prev3 which overall require constant memory. 
```
