```python
class Solution:
    def earlierstAcq(logs,n):
        #BOTTOM-UP MEMOIZATION MERGE SORT###########
        def merge(left,right):
            ptr1 = 0
            ptr2 = 0
            merged = [_ for i in range(len(left)+len(right))]
            
            for i in range(len(merged)):
                if (ptr1 >= len(left)) or ((ptr2 < len(right)) and left[ptr1][0] >= right[ptr2][0]):
                    ptr2 += 1
                    merged[i] = right[ptr2]
                elif (ptr2 >= len(right)) or (ptr1 < len(left) and left[ptr1][0] <= right[ptr2][0]):
                    ptr1 += 1
                    merged[i] = left[ptr1]
                    
            return merged
                    
            
        def sortArray(logs):
            if len(logs) <= 1:
                return logs
            
            mid = len(logs) //2
            n = len(logs)
            left = logs[0:mid]
            right = logs[mid:n]
            sorted_left = sorted(left)
            sorted_right = sorted(right)
            return merge(sorted_left,sorted_right)
        
        def find(vertex,root):
            if vertex == root[vertex]:
                return vertex
            
            root[vertex] = find(root[vertex],root)
            return root[vertex]
        
        def shouldUnion(x,y,root,rank):
            xroot = find(x,root)
            yroot = find(y,root)
            if xroot == yroot:
                return False
            if rank[xroot] > rank[yroot]:
                root[yroot] = xroot
            elif rank[yroot] > rank[xroot]:
                root[xroot] = yroot
            else:
                root[yroot] = xroot
                rank[xroot] += 1
            return True
            
        ############################
        
        sorted_logs = sortArray(logs)
        connected_components = n
        root = [0 for i in range(n)]
        rank = [1 for i in range(n)]
        
        for i in range(n):
            root[i] = i
        
        for log in sorted_logs:
            connected_components -= shouldUnion(log[1],log[2],root,rank)
            if connected_components == 1:
                return log[0]
        return -1
    
```

### Algorithm Explanation
Let's envision this problem as a graph where each connected component, or disjoint set represents a group of acquaintances. This means that when every person belongs to the same set, they are acquaintances of one another, so the moment we discover that there exists only one set left (i.e one root node), return the timestamp. In the graph, a direct edge between two nodes represents a friendship relation. However, if there is an indirect edge between two nodes (i.e there is still a path between the pair of nodes), then there is an acquaintanceship between both people. We know that in a fully-connected graph that for every pair of nodes, there is a path between them. So, the moment our graph is fully-connected, we are done since everyone is an acquaintance of one another. Algorithm-wise we will initialize the number of connected components to be N, or the second parameter. Every time, we establish an edge connection between two vertices that initially belonged to distinct sets, then we'll decrement the number of connected components. However, if the two vertices whose sets we wanted to perform the union operation on share the same root node, then we won't decrement the number of connected components. The omment we notice that the counter var for the number of connected components is equal to 1.   


### Bottom-Up Approach for Merge Sort
#### Divide-and-Conquer Approach:
Divide Step: Divide the problem into N subproblems where N >= 2. Split the input array into two equal sized subarrays (the first subarray will span the indices 0 to len(array) // 2 and the second subarray will be len(array) // 2 to len(array)) . We will continue to follow this divide step into we reach the smallest unit subarray that is already sorted which is a subarray of size 1. 

#### Conquer: 
Solve each subproblem recursively. After hitting the base case of a subarray of size 1, we will execute the recursive calls in LIFO order and start merging the subarrays into even bigger arrays. 

#### Combine:
Combine the results of each subproblem. Create an intermediate array with a total size that is the sum of the subarray sizes. Each of these subarrays will already be sorted because we've solved our subproblems in our conquer step. Maintain a pointer that will traverse through the first subarray ptr1 and a pointer that will traverse through the second subarray: ptr2. So, we will compare the entry of the first subarray at index ptr1 with the entry of the second subarray at index ptr2 and whichever entry is smaller will be placed into the next open merged array entry. Increment the pointer of the subarray whose element was picked and then repeat the process.

Runtime Analysis: We can envision our recursive calls as a recursive tree which will have a height of logn where N is the size of the original array we wanted to sort. This is because we have a branching factor of 2 for every recursive call as a result of invoking the recursive function once for the left subarray and then another time for the right subarray. If we were to raise 2 to the height of our recursion tree, which is log_2(N), we will get back N, which is the original number of array entries that we wanted to sort. We have log_2(N) levels of our recursion tree, and each recursive call requires an extra O(N) work to traverse each of the subarrays and then merge them into one auxiliary data structure, so the overall runtime of the merge sort algorithm is O(NlogN).

Space Complexity: We know that the height of our recursion tree is logN and we must go down a path of the recursion tree before hitting the base case and then executing the recursive calls in LIFO order. So our recursion stack will only ever need to store logN recursive calls. For each recursive call, we declare and later return an auxiliary data structure that represents the merged array for the two input subarrays. In the worst-case, this requires linear space for when we combine the results for the two subproblems of the original problem into a merged array with a size equal to N the original number of entries we wanted to sort. However, this extra linear space is only intermediately intertwined with the space and this memory can become recycled after its contents are merged into a larger array, so the upper bound on the amount of space we need is O(N), and overall memory usage is O(N).
