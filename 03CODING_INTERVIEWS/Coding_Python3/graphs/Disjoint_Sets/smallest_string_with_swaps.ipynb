```python
class Solution:
    def smallestStringWithSwaps(self, s: str, pairs: List[List[int]]) -> str:
        def find(vertex,root):
            if vertex == root[vertex]:
                return vertex
            root[vertex] = find(root[vertex])
            return root[vertex]
        
        def union(x,y,root,rank):
            rootx = find(x,root)
            rooty = find(y,root)
            if rootx == rooty:
                return
            if rank[rootx] > rank[rooty]:
                root[rooty] = rootx
            elif rank[rooty] > rank[rootx]:
                root[rootx] = rooty
            else:
                root[rooty] = rootx
                rank[rootx] +=1
                
                
        def mergesort(unsorted):
            if len(unsorted) <= 1:
                return unsorted
            midpoint = len(unsorted) //2
            left = mergesort(unsorted[0:midpoint])
            right = mergesort(unsorted[midpoint:])
            return merge(left,right)
            
        def merge(left,right):
            total_len = len(left) + len(right)
            merged = [0 for x in range(total_len)]
            left_ptr = 0
            right_ptr = 0
            
            for i in range(len(merged)): 
                if (left_ptr >= len(left)) or (right_ptr < len(right) and left[left_ptr] > right[right_ptr]):
                    merged[i] = right[right_ptr]
                    right_ptr +=1
                elif (right_ptr >= len(right)) or (left_ptr < len(left) and right[right_ptr] > left[left_ptr]):
                    merged[i] = left[left_ptr]
                    left_ptr +=1
                else:
                    merged[i] = left[left_ptr]
                    left_ptr += 1
            
            return merged
                
        
            
        
        output = str(s)
        root = [0 for x in range(len(s))]
        rank = [1 for x in range(len(s))]
        indices = []
        
        for i in range(len(s)):
            root[i] = i
        
        for p in pairs:
            union(p[0],p[1],root,rank)
        
     
        root_node = 0
        count = 0
        my_list = []
        d = dict()
        for i in range(len(root)):
            root_node = find(root[i],root)
            if root_node in d:
                d[root_node].append(i)
                my_list.append(i)
                count +=1
            else:
                indices = [i]
                d[root_node] = indices
        
        for i in range(len(my_list)):
            characters = []
            for i in range(count):
                characters.append(s[my_list[i]])
            
            characters = str(characters)
            
            sorted_indices = mergesort(my_list)
            sorted_chars = mergesort(characters)
            for i in range(len(sorted_indices)):
                output[sorted_indices[i]] = sorted_chars[i]
                
        return output
```

## Algorithm

### Disjoint Set Approach:
We will swap the characters at the pair of indices. We would like to swap then such that the individual characters being swapped will end up is alphabetical order within the  context of the entire string and for the entire string to be the lexicographically smallest string. We'll incorporate an underlying graph data structure where each vertex represents an index of the string s, and they initially are the root nodes of their own disjoint sets. For each pair that we'll need to swap from the vector of pairs, we'll perform union operation on their corresponding sets. By merging the sets containing each of the vertices together, the number of connected components will reduce by 1, and the indices of the characters we want will end up in the same set. Likewise, if another pair of indices contains one of the indices of a character that was already swapped, the other index in the pair will be merged into the same set after performing the union operation. We can use transitive property here: if vertex 1 swapped with vertex 2, and vertex 2 was swapped with vertex 3, then vertex 1 will be swapped with vertex 3, especially since the order of the swaps doesn't matter. This means that the characters of any of the vertices inside the same set can be swapped with one another. So,, let's sort the chracters corresponding to the indices in the same set after already having performed the union operation on each of the pairs to be swapped. Also, let's sort the indices (or the vertices inside the set). Now, we can re-assign the characters in alphabetical order to the available indices of the set in ascending order.Therefore, these characters will be in alphabetical order with respect to the swapped indices that belong to the same set and within the context of the entire alphabet string. 

#Find function with pass compression: The find function is recursive function which requires a base case and recurrence relation.

#Base Case: When the parent node of the current vertex is equal to the vertex itself, the current vertex is a root node. The purpose of the find function is to find the root node of the initial vertex input. So, return the root node (i.e current vertex at the base case).

#Recurrence Relation: What is the relationship between the result of the problem and the result of the subproblem? As we reach the root node of the initial input vertex, we will traverse the parent nodes of each vertex passed in to the recursive function. We know that these parent nodess will share the same root node as the initial input vertex, so we can assign the root nodes of each of these parent nodes to the return value from the call to the recursive function. Return this root node.

#Union function: We will implement the union by rank algorithm. First call the find function to figure out the root nodes for vertex x and vertex y (the first two parameters passed in to the union function). Now, compare the root nodes and make sure they are not already the exact same. Otherwise, vertex x and vertex y are already part of the same set, so no merging of sets is neccessary. Next, compare the ranks of the root nodes. The rank of the root node is the height of the tree that represents the set. If we want to minimize the height of the tree that represents the two sets we are performing the union operation on, we will want to make the tree of smaller height the subtree of the larger tree, so we want the root node of the higher rank to be the root node of the overall merged set. Therefore, if the rank of the root of vertex x is greater than the rank of y's root, then update the root of y's root to be the root of x. On the other hand, if the rank of the root of vertex y is greater than the rank of x's root, then update the root of x's root to be the root of y. If both ranks are equal, then it doesn't matter which root node is selected to be the root node of the merged set so long as the rank is properly updated. So, if the root of vertex x is chosen, then the rank of x's root should be incremented by 1.

#Inside the main function, first take a copy of the string for the output string so that we don't modify the original input. We will make a pass through the vector of pairs and call the union function of the pairs where the first index in the pair will be the first argument and the second vertex will be the second argument. After this, our root array will contain the parent nodes of each of the vertices that corresponding to the indices. Next, we will make a pass through the root array and call the find function on each entry to determine the root node for each vertex. The root node will become the key of the dictionary and the vertices inside the set rooted at the root node will be added to the list of the indices, which what the key is mapped to as a value for the dictionary. Afterwards, we have a hashmap whose keys are root nodes and whose values are the list of vertices inside the same set as the root node. Next, iterate through each key, val pair in the map. We need another list with the same size as the list of indices. Initialize the vector elements with the characters located at each index in the strings from the list of indices. Call the sort array function first on the list of indices and then next on the list of characters. Lastly, make a pass through one of the lists (both have the same size so it doesn't really matter which one and assign the ouput string at position index from the index list to the character from the character lists at the same position in both lists). 


#Top-Down Approach for Merge Sort O(NlogN):
#Divide: How do we divide the problem into a set of N subproblems where N >= 2? Split the array in half (2 subproblems) and sort the subarrays.
#Conquer: Recursively solve each subproblem. Call merge function on the left and right subarrays of the full array we want to sort (first two arguments). 
#Combine the results in the subproblems. Create a merged array whose size is the sum of the two subarray sizes. Then maintain a left pointer and right pointer that start at index 0 for the left subarray and the right subarray. Compare the entry at the left pointer for the left subarray with the entry at the right pointer for the right subarray, if it is leass than or equal to the entry at the right pointer, then add it to the next available index of the merged array and increment left pointer by 1. Otherwise, add the entry at the right pointer to the next available index of the merged array and increment the right pointer by 1. IF we run out of left subarray entries, use the right subarray entres and vice versa. 

#Runtime of Mergesort: O(NlogN) because we have a recursion tree with logN levels. Everytime we split our array/subarray in half, we have a branching factor of 2. If we raise 2 by the log2(N) power, we will get back the size of the array that we ant to sort, or N. At each level, we will merge two sorted subarrays together, which requires O(N) time to do a pass through the merged array and update its entries with the corresponding subarray entries in ascending alphabetical order. Therefore, the overall runtime for the logN levels and the extra O(N) work at each level would be O(NlogN).
